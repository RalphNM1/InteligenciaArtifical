```python
import numpy as np
import matplotlib.pyplot as plt
```

### Crear un gráfico simple
Especificamos valores para los dos ejes *x* e *y* y lo mostramos

```python
plt.plot([2, 4, 6, 8], 
         [4, 8, 12, 16])
plt.show()
```

#### Indicar el color de la línea

```python
plt.plot([2, 4, 6, 8], 
         [4, 8, 12, 16],
         color='red')
plt.show()
```

#### Indicar etiquetas para los ejes y fijar un color

```python
plt.plot([2, 4, 6, 8], 
         [4, 8, 12, 16])

plt.xlabel('x', fontsize=15, color='green')
plt.ylabel('2*x', fontsize=15, color='green')
plt.show()
```

#### Gráfica de la función seno
Las funciones trigonométricas las obtenemos con *numpy*


```python
x = np.linspace(start=0, stop=10, num=50)  #espaciando números uniformemente a lo largo del intervalo
```

```python
plt.plot(x, np.sin(x))
plt.xlabel('x', fontsize=15, color='green')
plt.ylabel('sin(x)', fontsize=15, color='green')
```

#### Formatear los ticks y sus etiquetas
Ejemplo de formateo de los ticks del eje y:
* Las marcas de los ticks en color rojo
* Las etiquetas de los ticks en azul
* El tamaño del texto de las etiquetas: xx-large

```python
plt.tick_params(axis='y',
                color='red',
                labelcolor='blue',
                labelsize='xx-large')
```

Si lo que quieres es modificar las etiquetas de los ticks utiliza xticks/yticks

```python
plt.tick_params(axis='y',
                color='red',
                labelcolor='blue',
                labelsize='xx-large')
plt.xticks([0,2,4,6,8,10])
plt.show()
```

#### Añadir una leyenda a la gráfica

Con legend tenemos una leyenda dentro del área del gráfico.

Si la queremos fuera o se tienen varias subgráficas podemos utilizar <a href="https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figlegend.html#matplotlib.pyplot.figlegend"> figlegend()</a>

```python
x = np.linspace(start=0, stop=10, num=50)
plt.plot(x, np.sin(x), label='función seno')
plt.legend() 
plt.show()
```

#### Poner un título a la gráfica

```python
plt.plot(x, np.sin(x), label='función seno')
plt.legend()
plt.title('Título de la gráfica')
plt.show()
```

#### Fijando límites a los ejes
xlim() e ylim(): para fijar límites al eje x e y, respectivamente 

```python
plt.plot(x, np.sin(x), label='Función seno')
plt.legend()
plt.title('Título de la gráfica')
plt.xlim(1,5)
```

#### Ahora añadimos límite al eje y

```python
plt.plot(x, np.sin(x), label='Función seno')

plt.legend()
plt.title('Título de la gráfica')

plt.xlim(1,5)
plt.ylim(-1,0.5)
```

#### Múltiples plots en un pyplot
Los colores de cada plot, si no se indica uno determinado, se eligen iterativamente sobre una paleta de colores, que por defecto es: {'tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple', 'tab:brown', 'tab:pink', 'tab:gray', 'tab:olive', 'tab:cyan'} 

```python
plt.plot(x, np.sin(x), label='seno')
plt.plot(x, np.cos(x), label='coseno')

plt.legend()
plt.title('Prueba con gráficas')

plt.show()
```

#### Especificando colores
En este caso se han elegido, verde y magenta:
* En el caso del verde se indica el nombre completo
* Magenta se especifica con el nombre corto ('g' sería para el verde) <br />

Puedes consultar la lista de colores y códigos aquí:
https://matplotlib.org/stable/api/colors_api.html


```python
plt.plot(x, np.sin(x), label='seno', color='green')
plt.plot(x, np.cos(x), label='coseno', color='m')

plt.legend()
plt.title('Título de la gráfica')

plt.show()
```

### Formatos de líneas y marcadores

Ejemplo de gráfica con línea. Estilos de línea más comunes: 

- '-' or 'solid'
- '--' or 'dashed'
- '-.' or 'dashdot'
- ':' or 'dotted'

Formatos de marcadores: https://matplotlib.org/stable/api/markers_api.html 

```python
plt.plot(x, np.sin(x), label='seno', color='green',linestyle="--", linewidth=1, marker='*' )
plt.plot(x, np.cos(x), label='coseno', color='m', linestyle=":", linewidth=1, marker='v')

plt.legend()
plt.title('Título de la gráfica')

plt.show()
```

## Figuras

Un objeto "figure" representa la figura en la que se muestran las gráficas.

Se puede configurar de múltiples formas, en el siguiente enlace se pueden consultar las distinas partes de una figura:
https://matplotlib.org/stable/users/explain/quick_start.html

Fíjate en los Axes: área donde se dibujan los datos, son "contenedores" donde se dibujan los gráficos. 

Se puede usar el parámetro figsize para fijar el *ancho, alto* de la figura en pulgadas:

```python
plt.figure(figsize=(10,5))
plt.plot(x, np.sin(x), label='seno', color='green',linestyle="--", linewidth=1, marker='*' )
plt.plot(x, np.cos(x), label='coseno', color='m', linestyle=":", linewidth=1, marker='v')
plt.legend()
plt.title('Título de la gráfica')
plt.show()
```

## Subplots (forma 1)

Podemos mostrar varias gráficas en una misma figura. Para ello, vamos a utilizar la función <code>subplot</code>(<a href="https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html#matplotlib.pyplot.subplot">documentación de la función</a>) con tres parámetros:
<ul>
    <li>Número de filas totales</li>
    <li>Número de columnas totales</li>
    <li>Índice de la gráfica dentro del número total de gráficas</li>
</ul>

```python
plt.figure(figsize=(20,5))
plt.subplot(1,2,1) # 1 - numrows, 2 - numcols, 1 - image index
plt.title('Función seno')
plt.plot(x, np.sin(x), label='seno', color='green',linestyle="--", linewidth=1, marker='*' )
plt.legend()
plt.subplot(1,2,2) # # 1 - numrows, 2 - numcols, 2 - image index
plt.title('Función coseno')
plt.plot(x, np.cos(x), label='coseno', color='m', linestyle=":", linewidth=1, marker='v')
plt.legend()
plt.show()
```

## Subplots (forma 2)

Otra forma de mostrar varias gráficas en una misma figura con la función <code>subplots</code>
(<a href="https://matplotlib.org/stable/api/_as_gen/matplotlib.figure.Figure.html?highlight=subplots#matplotlib.figure.Figure.subplots">documentación de la función</a>) con dos parámetros:
<ul>
    <li>Número de filas totales</li>
    <li>Número de columnas totales</li>
</ul>

```python
fig, axs = plt.subplots(1, 2)  # 1 - numrows, 2 - numcols
fig.set_figwidth(15)
fig.set_figheight(5)
axs[0].set_title('Función seno')
axs[0].plot(x, np.sin(x), label='seno', color='green',linestyle="--", linewidth=1, marker='*' )
axs[0].legend()
axs[1].set_title('Función coseno')
axs[1].plot(x, np.cos(x), label='coseno', color='m', linestyle=":", linewidth=1, marker='v')
axs[1].legend()
plt.show()
#observa las opciones de copiar, expandir o guardar si te acercas a la imagen dependiendo de tu entorno
```
