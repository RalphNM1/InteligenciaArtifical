Realizar un programa en Python que pida al usuario la dimensión y los elementos de dos listas de números enteros (pueden tener distinta dimensión):

- Una vez leídos los datos y almacenados en las listas, se generará una nueva lista que será la intersección de ambas, es decir, contendrá aquellos elementos que se encuentren en ambas listas.
- El vector intersección se mostrará por pantalla siempre que existan elementos comunes.

NOTA: Se comprobará que el usuario no introduzca elementos repetidos dentro del mismo vector.


#### Ejemplo de ejecución:
<pre>
Introduce el numero de elementos de la lista 1 (>0):  3
Introduce los números:
Dato 1: 2
Dato 2: 4
Dato 3: 1

Introduce el numero de elementos de la lista 2 (>0):  5
Introduce los números:
Dato 1: 3
Dato 2: 3
Error, elemento repetido. Dato 2: 2
Dato 3: 9
Dato 4: 0
Dato 5: 1

La intersección de las listas es: 2 1

-----------------------------------------------------------

Introduce el numero de elementos de la lista 1 (>0):  2
Introduce los números:
Dato 0: 3
Dato 1: -2

Introduce el numero de elementos de la lista 2 (>0):  3
Introduce los números:
Dato 0: 0
Dato 1: 4
Dato 2: 5

No hay elementos comunes
</pre>

```python
# Realiza tu solución aquí o en un archivo .py independiente


def introdcuir_dimension(x):
    dimension = int(input(f"Introduce el numero de elementos de la lista {x} (>0): "))

    while dimension <= 0:
        print("Error, la dimensión tiene que ser mayor a 0.")
        dimension = int(input(f"Introduce el numero de elementos de la lista {x} (>0): "))
    return dimension

def introducir_numeros_lista(dimension,lista = []):
    print("Introduce los números: ")
    for x in range(dimension):
        while True:
            print(f"Dato {x + 1}:", end=" ")
            numero = int(input())
            if numero in lista:
                print("Error, elemento repetido. ", end="")
            else:
                lista.append(numero)
                print(lista[x])
                break
    return lista

def interseccion(lista1, lista2):
    lista3 = [value for value in lista1 if value in lista2]
    return lista3


def main():

    lista1 = []
    lista2 = []

    lista1 = introducir_numeros_lista(introdcuir_dimension(1),lista1)

    print("Lista final:", lista1)

    lista2 = introducir_numeros_lista(introdcuir_dimension(2),lista2)

    print("Lista final:", lista2)


    n_repetidos_listas = interseccion(lista1,lista2)


    if (n_repetidos_listas == []):
        print("No hay elementos comunes")
    else:
        print(f"La intersección de las listas es: {n_repetidos_listas}")


main()

```

```python
# Realiza tu solución aquí o en un archivo .py independiente
def main():
    frase = str(input("Dime una frase: "))
    caracter = input("Dime un caracter: ")
    fmodificado = ""
    for c in frase:
        if c in 'aeiouAEIOU':
            fmodificado += caracter
        else: 
            fmodificado += c

    print (f"La nueva frase es: {fmodificado}") 


main()



```

Realizar un programa en Python que permita gestionar las subscripciones de un usuario a diversas plataformas de entretenimiento, utilizando para ello **un diccionario**.

- El ejercicio parte de un diccionario que contiene el nombre y precio mensual de las subscripciones, formado por la clave (nombre de la plataforma) y el valor (precio mensual):


<pre>
        plataformas = {
        "Netflix": 11.99,
        "HBO": 8.99,       
        "Disney+": 7,
        "DAZN": 9.99,
        "Spotify": 9.99,
        "Apple Music": 9.99,
        "Amazon Prime": 3.99,
        }
</pre>


A continuación, deberá mostrar repetidamente un menú con las siguientes 6 opciones: 

1. Añadir plataforma (permite añadir una nueva plataforma al diccionario)
2. Eliminar plataforma (elimina una plataforma del diccionario)
3. Mostrar plataforma (muestra el precio de una plataforma a partir de su nombre)
4. Listar plataformas (muestra la lista de plataformas y sus precios)
5. Subscribirse n meses a una plataforma (muestra el precio total de la subscripción) 
6. Terminar (sale del programa)



#### Ejemplo de ejecución:

<img src="img/imagen1.png">

```python
# Realiza tu solución aquí o en un archivo .py independiente
salir = False
plataformas = {
    "Netflix": 11.99,
    "HBO": 8.99,       
    "Disney+": 7,
    "DAZN": 9.99,
    "Spotify": 9.99,
    "Apple Music": 9.99,
    "Amazon Prime": 3.99,
}

while not salir:
    print("\nMenú:")
    print("1. Añadir plataforma (permite añadir una nueva plataforma al diccionario)")
    print("2. Eliminar plataforma (elimina una plataforma del diccionario)")
    print("3. Mostrar plataforma (muestra el precio de una plataforma a partir de su nombre)")
    print("4. Listar plataformas (muestra la lista de plataformas y sus precios)")
    print("5. Subscribirse n meses a una plataforma (muestra el precio total de la subscripción)")
    print("6. Terminar (sale del programa)")

    opcion = input("Introduzca la opción (1-6): ")
    if opcion.isdigit():
        opcion = int(opcion)
    else:
        print("Por favor, introduzca un número válido.")
        continue

    if opcion == 1:
        plataforma_nombre = input("Introduzca el nombre de la plataforma: ")
        plataforma_precio = input("Introduzca el precio de la plataforma: ")
        if plataforma_precio.replace('.', '', 1).isdigit():
            plataformas[plataforma_nombre] = float(plataforma_precio)
            print(f"Plataforma {plataforma_nombre} añadida con éxito.")
        else:
            print("El precio debe ser un número válido.")

    elif opcion == 2:
        plataforma_nombre = input("Introduzca el nombre de la plataforma a eliminar: ")
        if plataforma_nombre in plataformas:
            del plataformas[plataforma_nombre]
            print(f"Plataforma {plataforma_nombre} eliminada con éxito.")
        else:
            print("La plataforma no existe en el diccionario.")

    elif opcion == 3:
        plataforma_nombre = input("Introduzca el nombre de la plataforma: ")
        if plataforma_nombre in plataformas:
            print(f"El precio mensual de {plataforma_nombre} es {plataformas[plataforma_nombre]} euros.")
        else:
            print("La plataforma no existe en el diccionario.")

    elif opcion == 4:
        print("Lista de plataformas y precios:")
        for nombre, precio in plataformas.items():
            print(f"- {nombre}: {precio} euros")

    elif opcion == 5:
        plataforma_nombre = input("Introduzca el nombre de la plataforma: ")
        if plataforma_nombre in plataformas:
            meses = input("Introduzca el número de meses: ")
            if meses.isdigit():
                total = plataformas[plataforma_nombre] * int(meses)
                print(f"El precio total por {meses} meses de {plataforma_nombre} es {total} euros.")
            else:
                print("El número de meses debe ser un entero válido.")
        else:
            print("La plataforma no existe en el diccionario.")

    elif opcion == 6:
        salir = True
        print("Gracias por usar el programa. Hasta pronto!")

    else:
        print("Opción no válida. Por favor, elija una opción del 1 al 6.")
```
